## Hombres lobo

### En el país de los hombres lobo sucede que los humanos **siempre** dicen la verdad y los hombres lobo **siempre** dicen mentiras. 

#### 1. Estando en el país de los hombres lobo te encuentras a tres lugareños, cada uno hace una aseveración:
+ Antonio dice: **Al menos uno de nosotros es un hombre lobo**.
+ Barbara dice: **Al menos uno de nosotros es humano**.
+ Catalina dice: **Exactamente dos de nosotros son hombres lobo**.

*¿Quién dice la verdad?*

#### 2. Más adelante, en un café encontraste a otros 4 aldeanos, los cuales te dicen que:
+ Dario: **Eleonora es humana**.
+ Eleonora: **Fátima es humana**.
+ Fátima: **Dario es humao**.
+ Germán: **Como máximo, dos de nosotros somos humanos**.

Al siguiente día te enteras que Germán apareció muerto, devorado por hombres lobo y sabes que los hombres lobo no comen hombres lobo, de modo que Germán era claramente un humano.

*¿Quienes de Dario, Eleonora y Fátima son hombres-lobo?*

#### 3. En el tumulto del asesinato, contrario a toda lógica, en vez de salir corriendo de un pueblo de hombres lobos, te ganó tu lado *curioso* y te pusiste a platicar con las personas que estaban ahí.  En total había 12 personas, las cuales te dijeron:

+ Persona 1: **La persona 2 es un hombre-lobo**.
+ Persona 2: **La persona 3 es un hombre-lobo**.
+ Persona 3: **La persona 4 es un hombre-lobo**.
...
+ Persona 12: **Como máximo entre nosotros hay 6 hombres lobo**.

*¿Qué es la persona 1?**

#### 4. Ya de salida de dicho país, cuando ya te habías aburrido de los humanos y los hombres lobo te enteras que hay otro tipo de aldeano.  Lo llaman un *colaborador* y es un humano que trabaja con los hombres lobo y siempre dice mentiras.  Sabes que hay un *colaborador* en el hotel dónde te encuentras.  Hay cuatro personas y cada una te dice lo siguiente:
+ Abdón: **Exactamente uno de nosotros es humano. Corina es la colaboradora**.
+ Beatriz: **Exactamente dos de nosotros son humanos. Yo no soy la colaboradora**.
+ Corina: **Exactamente tres de nosotros son humanos. Damián es el colaborador**.
+ Damián: **Exactamente cuatro de nosotros son humanos. Beatriz es la colaboradora**.

*¿Quién es el colaborador?*

#### 5. Cuatro alumnos, Abel, Brenda, Cristian y Diana toman la misma clase.  Dos de ellos son derechos y dos son zurdos. Un día el profesor les aplicó un examen en el que obtuvieron 250, 200, 150 y 100 puntos (en algún orden). Basándote en las aseveraciones siguientes...

+ Aunque el alumno con más puntos puede haber sido diestro o zurdo, cuando se enlista a los 4 en orden del puntaje más alto al más bajo, los zurdos y los diestros quedan alternados.
+ Abel es diestro.
+ Diana obtuvo mejor puntaje que Brenda.
+ Cristian obtuvo mejor puntaje que los dos alumnos diestros.

*¿Qué alumnos obtuvieron 200 y 150 puntos respectivamente?*

#### 6.Elena, Fernando y Gabriela siguen comiendo galletas (cooperando para que México sea el primer lugar mundial en obesidad infantil).  En su último concurso quedaron en 1er, 2do y 3er lugar en algún orden.  Además esta vez, cada uno de ellos participó comiendo un tipo de galleta distinta.  Las galletas que comieron fueron Oreo, Emperador y Arcoiris.  Basándote en las aseveraciones siguientes...

+ Gabriela y el concursante que comió *oreos* quedaron juntos en lugar.
+ Elena no comió *oreos*.
+ El concursante que comió *emperador* le ganó a Gabriela.
+ El concursante que comió *oreo* quedó junto del concursante que comió *emperador*.

*¿En qué orden quedaron?*

#### 7. Cuatro amigos, David, Erika, Fabio y Gerardo participaron en una carrera de bicicletas quedando en 1ro, 2do, 3er y 4to lugar en algún orden. Cada uno de ellos usó una bici de color distinto. Basándote en las aseveraciones siguientes...

+ Una de las bicicletas era roja.
+ Erika usó una bicicleta morada. 
+ Fabio **no** usó la bicicleta verde.
+ David y Erika no fueron ni primero ni último.
+ El concursante de la bicicleta azul terminó antes que Gerardo.
+ El concursante de la bicicleta verde le ganó al concursante de la bicicleta morada.

*¿En qué orden quedaron y qué bicicleta usó cada quien?*

####  8. Tres alumnos del Carol (de quienes no diremos nombre) son áltamente bromistas.  Estos alumnos, a los que llamaremos A, B y C todos los días de clase hacen alguna broma.  Las bromas siempre suceden al inicio de clases, durante el primer recreo y a la hora de la salida.  El día de hoy las bromas fueron: una bomba fétida, pintura fresca en la pared de la escalera y papel de baño con pica-pica.  A, B y C tuvieron además el cinismo de dejar pistas, pero entre las pistas que dejaron, avisaron que una de ellas es falsa.  Basándote en las pistas...

+ *A* hizo su broma después de la broma de la pintura fresca.
+ La broma de la pintura fresca sucedió después de que *B* hiciera su broma.
+ La broma de *B* sucedió después de la broma de la bomba fétida.
+ *C* hizo la broma de la bomba fétida.
+ Las bromas del papel de baño y la bomba fétida ocurrieron una después de la otra (en algún orden).

*¿Qué broma hizo cada quien y en qué orden las ejecutaron?*
**NOTA**: Recuerda que una de las pistas es falsa.


## Información perfecta

#### 9. Joshua y Sebastián quieren saber la fecha de cumpleaños de Ana Sofía.  A ella no le causa tanta emoción compartírselas así que decide decirle a Joshua el mes en el que nació (sin que Sebastián oiga) y a Sebastián el día (sin que Joshua oiga), con la promesa de que ninguno compartirá la información con el otro.  Como la información no es suficiente, le piden a Roberto una pista más, Roberto acepta darles una lista de posibles fechas.  La lista es la siguiente:
+ Mayo 15, mayo 16, mayo 19, junio 17, junio 18, julio 14, julio 16, agosto 14, agosto 15 o agosto 17.

Viendo esta lista sucede el siguiente diálogo entre Joshua y Sebastián:

+ Joshua: **No sé cuando sea el cumpleaños de Ana, pero estoy seguro de que Sebastián tampoco sabe**.
+ Sebastián: **Al principio no sabía cuando era, pero ahora ya sé**.
+ Joshua: **Ahh, entonces yo también ya sé**.

Contentos ambos, fueron a comprar un regalo para el cumpleaños de Ana Sofi.  

*¿Cuándo cumple años Ana Sofía?*

#### 10. Camino al Sanborns a comprar el regalo, Sebastián y Joshua dieron una vuelta equivocada y entraron a la *dimensión desconocida*, ¿o era ahora el mundo de Stranger Things?, sea como sea, terminaron atrapados en la cárcel de un carcelero sádico.  El carcelero los puso a cada uno en una celda distinta.  Cada celda tiene árboles dentro (es un carcelero muy ecológico) y cada prisionero puede ver únicamente los árboles de su celda.  Todos los días el carcelero pasa con Joshua y le pregunta: *¿Si sumas los árboles de las dos celdas hay 18 o hay 20 árboles?* (el carcelero les asegura que una de las dos respuestas es correcta).  Joshua puede contestar o decidir *pasar*, en el caso de que *pase* el carcelero le hará la misma pregunta a Sebastián y también puede elegir contestar o *pasar*.   Si alguno de los dos contesta correctamente en algún momento, ambos serán liberados y regresados a su dimensión.  Si en algún momento uno de los dos contesta de forma incorrecta, ambos permanecerán por siempre convertidos en *umpa-lumpas*.

*¿Es posible que Sebastián y Joshua puedan salir y comprar el regalo por el que iban? Si sí ¿cómo? Si no ¿por qué?
