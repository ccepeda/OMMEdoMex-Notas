# Arte de las matemáticas (Carol Baur 2018-2019)

### Ejercicios y Tarea 06/09/2018

#### 1. Durante una mudanza algunas de tus cosas fueron empacadas en 4 cajas distintas.  A cada caja se le puso uno de los siguientes letreros: *calcetines*, *camisas*, *pantalones*, *camisetas*. Sabes que **Sólo uno** de los letreros es correcto, pero no recuerdas cuál es. 

*¿Cuál es el **menor** número de cajas que tienes que abrir para garantizar (sin importar la suerte) que puedes decir cuál de las cajas es la que contiene el letrero correcto?*

#### 2. En una de tus vidas anteriores fuiste pirata. En esa ocasión, te presentaron el siguiente acertijo. Te pusieron frente a 3 cofres cerrados que tenían las siguientes leyendas:

+ El primer cofre dice: En este cofre hay 100 monedas de oro.
+ El segundo cofre dice: En este cofre hay 100 monedas de plata.
+ El tercer cofre dice: En este cofre hay 50 monedas de oro y 50 de plata.

Conociendo a los piratas, sabías que **TODOS** los letreros estaban mal puestos.  El capitán del barco te dijo que podías sacar una moneda, sin ver, del cofre que eligieras. Posterior a eso, deberías escoger un cofre.  Si elegías el cofre con 100 monedas de oro, era tuyo y podías irte.  Si eligías otro, morías. Tu yo de la vida anterior te contactó en un sueño para que le ayudes. 

*¿Qué instrucciones le darás?*

#### 3. Cinco hermanos, Alex, Bruno, Carlos, Daniel y Eduardo están comparando sus alturas y ven que:

+ Alex es más alto que Bruno pero más bajo que Carlos.
+ Daniel es más alto que Eduardo pero más bajo que Alex.

*¿Quién es el más alto de todos?*

#### 4. Ana, Beto y Caro participan en una carrera y terminan 1ro, 2do y 3ero (en algún orden).  Sabes qué:

+ Los lugares en los que quedaron Beto y Caro **no** son consecutivos.
+ Ana le ganó a Beto.

*¿Quién ganó la carrera?*

#### 5. Andrés, Blanca, Clara y David son concursantes de un *reality show*.  Ellos quedaron en 1ro, 2do, 3er y 4to lugar en algún orden.  Sabes que:

+ Blanca no fue primer lugar.
+ David no fue último lugar.
+ Clara no quedó ni junto de Blanca ni junto de David.
+ Andrés quedó en mejor lugar que Blanca.

*¿Quienes quedaron en segundo y tercer lugar?*

#### 6. Tres niños Elena, Fernando y Gabriela participan en un concurso para ver quién come más galletas.  Si Elena come más galletas que Fernando. 

*¿En cuantas formas distintas pueden haber quedado los lugares del concurso?*

#### 7. Angel, Brenda, Cori y Davino presentaron un examen en su escuela. En el examen, todas las preguntas se contestan con un verdadero o un falso.  El examen tenía 4 preguntas y la respuesta correcta a todas las preguntas era "verdadero".  Sabes que:

+ Los cuatro alumnos tuvieron calificaciones diferentes.
+ Ángel obtuvo la mayor calificación. 

*¿De cuántas formas distintas pudo Ángel haber contestado el examen?*

#### 8. De nuevo Elena, Fernando y Gabriela participan en el concurso de comer galletas, pero esta vez hay 10 niños compitiendo.  Sabes que:

+ Al final del concurso quedaron 7 niños entre Fernando y Gabriela
+ Al final del concurso quedaron 3 niños entre Elena y Gabriela

*¿En cuántos y cuáles lugares distintos pudo haber quedado Elena?*

#### 9. Hay 4 competidores en una carrera.  Supón que te dan pistas de la forma *"El competidor **X** terminó antes que el competidor **Y**"* 

+ *¿Cuál es el mínimo número de pistas que necesitas para conocer el orden final de todos los competidores?*
+ *Si en vez de 4 competidores son 6. ¿Cuántas pistas necesitarías?*
+ *¿Y si en vez de 6 competidores son **N** (donde **N** puede ser cualquier entero positivo)?*

#### 10. Andrés, Blanca, Clara y David regresan a la segunda temporada de su *reality show*.  Si sabes que;

+ El lugar en el que quedó Andrés no está junto del lugar en el que quedó Blanca.

*¿De cuántas formas distintas pueden haber quedado ordenados los participantes?


#### 11. Uno de tus compañeros, cansado de la escuela, decidió mejor ganarse la vida jugando poker.  En su primer torneo llegó a los últimos 5 competidores.  Si sabes que:

+ Tu compañero no quedó en primer lugar

*¿De cuántas formas distintas pudieron haber quedado los lugares de los úlitmos 5 competidores?

**TIP:** En este caso es más sencillo contar todas las formas en las que tu compañero pudo haber quedado en primer lugar y luego restarlas de todas la formas posibles en las que pudo haber terminado la competencia.


#### 12. Tienes 4 cartas de una baraja.  La reina de corazones, el rey de corazones, el 10 de tréboles y el 5 de tréboes.

*¿De cuántas formas distintas puedes acomodarlas en una fila de modo que las dos cartas de corazones **NO** queden juntas?


#### 13. De nuevo tu yo pirata tiene una pregunta para ti.  Esta vez está frente a dos cofres, sabe que uno de ellos tiene un tesoro y el otro una trampa mortal. Cada cofre tiene un letrero y sabes que **uno** o **ambos** de los letreros está mintiendo.  Los letreros dicen:

+ Cofre 1: *Este cofre tiene el tesoro*
+ Cofre 2: *Este cofre no tiene el tesoro*

*¿Qué cofre debe abrir?*


#### 14. En tu visita por el laboratorio de pociones del parque temático de Hogwarts encontraste un mistorioso recipiente con una confusa etiqueta.  El guía de turistas te dice que el contenido del recipiente puede ser un mortal veneno o una poderosa posicon de invencibilidad, pero que nadie sabe que tiene.  Lo que saben es que la leyenda de la etiqueta es **FALSA**.  La etiqueta dice lo siguiente:

+ *No es verdad que esto no es veneno.*

*¿Qué debes hacer? ¿Tomar la poción o dejarla para el siguiente turista incauto?* Recuerda que lo que dice la etiqueta es **falso**.

#### 15. En una de esas situaciones, que como dijera la *coach* de Gumball, suelen presentarse en la vida cotidiana.  Te encuentras, tras dar unas cuantas volteretas, huyendo de perros enojados, en un callejón y frente a 3 puertas.  Sabes que una de las puertas te lleva a la salvación y las otras dos a pozos sin fondo donde caerás por toda la eternidad.  Las puertas tienen letreros y de alguna forma sabes (eso es lo raro del suceso) que la puerta que te lleva a la salvación es la puerta que tiene un letrero falso, mientras que aquellas con letreros verdaderos te llevarán a los pozos sin fondo.  Las puertas dicen:

+ Puerta 1: La salvación está tras de la puerta 2.
+ Puerta 2: La salvación está tras de la puerta 3.
+ Puerta 3: La salvación no está tras de esta puerta.

*¿Qué puerta debes tomar?* Recuerda que la salvación está tras de la puerta con un letrero que sea falso.


#### 16. Un valioso collar fue robado.  El culpable es uno de tres famosos ladrones, Fantomas, El ministro D..., Chucho el Roto.  Los tres son amigos y los tres saben quién fue el que en verdad robo el collar.  El inspector de policía interrogó a los tres y obtuvo los siguientes testimonios:

+ Fantomas: El ministro D... robó el collar.  Yo nunca mentiría.
+ El ministro D...: Chucho el roto robó el collar.  Yo no robé el collar.
+ Chucho el roto: Fantomas robó el collar. Yo no robé el collar.

El inspector está perplejo ya que no encuentra la solución. Sabe, sin embargo, que cualquiera de los interrogados que esté diciendo la verdad, lo hace en ambas afirmaciones. Mientras que cualquiera de ellos que esté mintiendo, lo hace en ambas afirmaciones.

Hay una cuantiosa recompensa por recuperar el collar que compartirán contigo si ayudas a resolver el problema.

*¿Quién robó el collar?*

#### 17. La tercera, no sabemos si la última, temporada del *reality* de Andrés, Blanca, Clara y David ya está en Netflix.  Un *buen* amigo te *spoileo* parte del final y te dijo que:

+ Blanca terminó en mejor lugar que David y que Clara quedó en mejor lugar que Andrés.

Como te enojaste por ese *spoiler* tu amigo quizo redimirse haciendo el *spoiler* más interesante. Además de la información que ya te dió, te da 3 pistas y te dice que exactamente una de las pistas es falsa.  Las pistas son:

+ David quedó en un lugar junto del de Andrés.
+ Clara **no** quedó en un lugar junto al de Andrés.
+ El lugar en que quedó Blanca no está junto del que obtuvo Clara.

*Basado en estas pistas ¿Quiénes quedaron en tercer y cuarto lugar?
