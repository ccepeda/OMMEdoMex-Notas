## El emparedado de Esteban

#### Luego de aprender el juego del _dulce de ajo_ y de ver el famoso capítulo del emparedado de Homero (https://www.youtube.com/watch?v=Ep7217exFPw), Esteban decidió inventar su propio juego.  El _emparedado de Esteban_. Y se juega así:  Todos los lunes Esteban lleva un emparedado gigante a la escuela y lo guarda en su casillero, dónde permanece durante los siguientes 7 días.  El siguiente lunes (antes de meter el emparedado que lleva), Esteban se reune con un retador y sacan el emparedado viejo. A este emparedado le ensartan dos palillos chinos y en ambos palillos alguna cantidad de aceitunas.  Entonces, alternando turnos, cada jugador puede hacer su movimiento el cual consta de comer tantas aceitunas como quiera, pero con la restricción de que todas las aceitunas las debe tomar del mismo palillo.  Si ambos palillos están vacíos (sin aceitunas) cuando sea el turno de un jugador, ese jugador debe comerse el emparedado. Obviamente, dicho jugador se considera el perdedor.

*Ayúdale a Esteban a desarrollar una estrategia ganadora. Ya que lo hagas, no lo vayas a retar los lunes siguientes*

#### 1. En su primer juego, pusieron 4 aceitunas en el primer palillo y 1 aceituna en el segundo palillo. Si a Esteban le toca empezar...

*¿Qué estrategia deberá seguir Esteban para ganar?*

#### 2. Para el segundo juego había 11 aceitunas en el primer palillo y 9 en el segundo. De nuevo a Esteban le toca empezar...

*¿Qué esterategia debe seguir para ganar?*

#### 3. Dibuja la tabla de posiciones ganadoras y perderdoras para el juego de Esteban. (Llega hasta 8 x 8 posiciones). Generaliza en qué situaciones tendrá Esteban una estrategia ganadora siendo el primer jugador y en qué situaciones tendrá una estrategia ganadora siendo el segundo jugador.


## Alitas y cerveza

#### Después de jugar un partido de fútbol, aprovechando que ya son mayores de edad, Santiago y Sebastián se fueron a comer a la B.T.K. Satélite alitas y cerveza. A Sebastián le gustan las alitas *Atomic* y a Santiago las *Fresa Hot*, de modo que le pidieron al mesero que les trajera de esos dos tipos de alitas junto con una jarra de cerveza. En eso, se dieron cuenta de que no traían dinero suficiente así que acordaron que uno de los dos se sacrificará y se quedará a ver cómo paga la cuenta mientras el otro regresa tranquilamente a su casa y sigue con su vida normal.  Para decidir quién se sacrifica diseñaron un juego que se jugará por turnos. En su turno cada quien puede hacer una y sól una de las siguientes 4 cosas:
+ Comer alguna cantidad (mayor a 0), siempre y cuando queden, de alitas *Atomic*.
+ Comer alguna cantidad (mayor a 0), siempre y cuando queden, de alitas *Fresa Hot*.
+ Comer de ambos tipos de alitas alguna cantidad (mayor a 0). Pero la misma cantidad de ambos tipos.
+ Tomarse de *Hidalgo* (para una definición ver: https://verne.elpais.com/verne/2017/01/28/mexico/1485571093_004257.html) la jarra de cerveza.

#### Como ambos son buenos matemáticos saben que el juego tiene una estrategia ganadora.  Entonces, para hacerlo justo una vez que les trajeron las alitas y la jarra, te marcaron para que tú decidas quién tiene el primer turno.  ¿A quién quieres salvar?

#### 4. Si el mesero trajo 1 alita *Atomic*, 1 alita *Fresa Hot* y la jarra de cerveza...

*¿Qué jugador ganará el juego, el primero o el segundo?*

#### 5. Si el mesero trajo 5 alitas *Atomic*, 5 alitas *Fresa Hot* y la jarra de cerveza...

*¿Qué debe hacer el primer jugador en su primer turno para ganar?*

#### 6. En la B.T.K. puedes pedir tantas alitas como puedas comer (asumiendo claro que tienes la paciencia de esperar por ellas, ya que suelen tardar mucho). Si modifican el juego para que el segundo jugador pueda pedir de un tipo de alitas antes de que empiece el juego y al inicio el mesero trajo 7 alitas *Atomic* y 6 alitas *Fresa Hot*....

*¿Cuántas alitas debe pedir el jugador dos para poder ganar el juego?*

#### 7. Dibuja y describe cómo sería una forma de llevar la tabla de posiciones ganadores y perdedoras para este juego.

## Paulina y sus vasos

#### Durante su larga trayectoria en concursos de matemáticas y programación Paulina aprendió un juego con el que gusta de apantallar a los incautos.  Este juego se realiza con 3 vasos comunes y corrientes. Los vasos se colocan sobre una mesa, se pueden poner con la *boca* hacia arriba o hacia abajo.  El objetivo del juego es dejar todos los vasos con la *boca* hacia arriba.  Para lograrlo el jugador puede tomar cualesquiera dos vasos y voltearlos. Si en un número finito de movimientos el jugador logra dejar todos los vasos con la *boca* hacia arriba, habrá ganado.  Si quieres ver una versión escenificada de Paulina ganando el juego a un *incauto* puedes checarlo aquí: https://www.youtube.com/watch?v=lWUzFuesgMk

#### 8. En el video vemos que Paulina inicia el juego con la configuración D - U - D.  En este caso usamos U para indicar que el baso quedó boca arriba (Up) y D para indicar que el vaso quedó boca abajo (Down).  Y siempre logra ganar, mientras que su contrincante inicia con la configuración U - D - U y nunca lo logra.

*Lista todas las posibles configuraciones iniciales con las que es posible ganar el juego.  

#### 9. Si ahora cambiamos la regla a que en vez de poder voltear cualesquiera 2 vasos tengas que voltear dos vasos que están juntos....

*¿Cuáles son todas las configuraciones de inicio que son ganadoras?*

#### 10. Si en vez de ser 3 vasos, fueran 7 y se mantiene la regla de que sólo puedes voltear dos vasos que estén juntos...

*¿Cuáles de estas configuraciones iniciales son ganadoras y cuáles perdedoras?*

+ U D U D U D U
+ D U D U D U D
+ D U D D D U D
+ U D U U U D U
+ D D U U D D U
+ U U D U U D D

#### 11. Si en vez de 7 son *N* vasos y sólo puedes voltear monedas que estén juntas....

*¿Cuál es la condición necesaria para asegurar que una configuración inicial sea ganadora?*
*¿Puedes demostrar que tu aseveración es cierta?*
