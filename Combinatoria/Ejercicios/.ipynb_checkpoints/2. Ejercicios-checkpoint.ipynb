#### 2.1
¿De cuántas formas distintas se pueden sentar en una mesa circular 5 niños y 5 niñas? si...

+ a) No hay ninguna restricción.
+ b) Los 5 niños deben estar juntos.
+ c) Cada niño debe ir sentado junto a 2 niñas.

#### 2.6

#### 2.7

#### 2.19

#### 2.20

#### 2.21
