# Notas de combinatoria - Resumen
### OMM Estado de México

El siguiente es un resumen de los puntos cubiertos en las notas.

### Principios de conteo

#### 1. Principio de la adición
Si $A_1, A_2, ... , A_n$ son conjuntos finitos disjuntos, entonces $|A_1 \cup A_2 \cup ... \cup A_n| = |A_1| + |A_2| + ... + |A_n|$

#### 2. Principio de la substracción
Si $B \subseteq A$, entonces $|A - B| = |A| - |B|$
    
#### 3. Principio de multiplicacion *(Prinicipio fundamental del conteo)*
Si $A_1, A_2, ... , A_n$ son conjuntos finitos, entonces el número de *n*-tuplas $(a_1, a_2, ... , a_n)$ que satisfacen que $a_i \in A_i$ para toda $i$ es $|A_1| \times |A_2| \times ... \times |A_n|$

#### 4. Principo de la división
Si $T$ y $S$ son conjuntos finitos tales que existe una relación $d$-a-uno de $f:T \to S$, entonces $|S|=\frac{|T|}{d}$


### Fórmulas de enumeración

#### 5. El número de palabras de longitud $k$ sobre un alfabeto de $n$ símbolos es $n^k$.

#### 6. El número de formas de *acomodar* $n$ objetos distintos en una línea es $n!$.

#### 7. El número de formas de *acomodar* $k$ objetos obtenidos de un conjunto original de $n$ objetos distintos en una línea, con $k \leq n$ es $n \times (n-1) \times (n-2) \times ... \times (n-k+1)$. También puede escribirse como $\frac{n!}{(n - k)!}$

#### 8. El número de subconjuntos distintos de tamaño $k$ obenidos a partir de conjunto de tamaño $n$ con $k \leq n$ es $\binom{n}{k} = \frac{n!}{k! (n-k)!}$ y se conoce como coeficiente binomial.  

#### 9. El número de formas distintas de acomodar $n$ objetos, $a_i$ de los cuales son de tipo $i$ en una línea es $\binom{n}{a_1, a_2, ... , a_k} = \frac{n!}{a_1! a_2! ... a_k!}$ y se conoce como coeficiente multinomial.  

#### 10. *(CAMINOS)* El número de multiconjuntos de $n$ elementos formados a partir con $k$ tipos de elementos distintos es $\binom{n+k-1}{n}$.  

#### 11. La cantidad de composiciones débiles de $n$ en $k$ partes es $\binom{n+k-1}{n}$.

#### 12. La cantidad de composiciones de $n$ en $k$ partes es $\binom{n-1}{k-1}$.

#### 13. Teorema del binomio
Si $n$ es un entero positivo entonces
<center>$(x + y)^n = \sum_{k=0}^{n} \binom{n}{k} x^k y^{n-k}$</center>


### Identidades importantes

#### $\binom{n}{k} = \binom{n}{n - k}$

#### $\binom{n}{k} + \binom{n}{k + 1} = \binom{n+1}{k+1}$

#### $\sum_{k=0}^{n} \binom{n}{k} = 2^n$

#### Para $n$ impar $\binom{n}{0} + \binom{n}{2} + ... + \binom{n}{n-1} = \binom{n}{1} + \binom{n}{3} + ... + \binom{n}{n}$

#### Para $n$ par $\binom{n}{0} + \binom{n}{2} + ... + \binom{n}{n} = \binom{n}{1} + \binom{n}{3} + ... + \binom{n}{n-1}$


### Otros hechos importantes

#### Principio de las casillas
Si $A_1, A_2, ... , A_k$ son conjuntos finitos disjuntos dos-a-dos tales que $|A_1 \cup A_2 \cup .... \cup A_k| > kr$ entonces existe al menos un índice $i$ tal que $|A_i| > r$

#### Principio de inclusión-exclusión
Sean $A_1, A_2, ... , A_n$ conjuntos finitos.  Entonces 
<center> $|A_1 \cup A_2 \cup ... \cup A_n| = \sum_{j=1}^{n} (-1)^{j-1} \sum_{i_1, i_2, ..., i_j}^{} |A_{i_1} \cap A_{i_2} \cap ... \cap A_{i_j}|$ </center>

donde $(i_1, i_2, ..., i_j)$ son todos los subcojuntos de $j$ elementos de $n$.

Por ejemplo, si $n = 2$ entonces

<center>$|A_1 \cup A_2| = |A_1| + |A_2| - |A_1 \cap A_2|$</center>

si $n = 3$ entonces

<center>$|A_1 \cup A_2 \cup A_3| = |A_1| + |A_2| + |A_3| - |A_1 \cap A_2| - |A_1 \cap A_3| - |A_2 \cap A_3| + |A_1 \cap A_2 \cap A_3|$</center>
